```python
import numpy as np 
import pandas  as pd
import matplotlib.pyplot as plt
from sklearn.tree import DecisionTreeClassifier
```

```python
def analyzer(max_depth):
    df = pd.read_csv("Social_Network_Ads.csv")
    df['Gender'].replace ({"Male" :0, "Female": 1}, inplace = True)
    x = df.iloc[:, 2:4].values
    y = df.iloc[:, -1].values
    
    
    clf1 = DecisionTreeClassifier(max_depth=max_depth)
    clf1.fit (x, y)
    
    a = np.arange(start=x[:, 0].min() - 1, stop=x[:, 0].max() + 1, step=0.1)
    b = np.arange(start=x[:, 1].min() - 1, stop=x[:, 1].max() + 1, step=100)

   # a = np.arrange(start = x[:,0].min()-1, stop = x[:,0].max() +1, step = 0.1)
   # b = np.arrange(start = x[:,1].min()-1, stop = x[:,1].max() +1, step = 100)
    
    xx, yy = np.meshgrid(a,b)
    
    input_array = np.array([xx.ravel(), yy.ravel()]).T
    
    labels = clf1.predict(input_array)
    
    plt.contour(xx, yy, labels.reshape(xx.shape), alpha = 0.5)
    plt.scatter(x[:,0], x[:,1], c = y)
    
```

```python
analyzer(max_depth=None)
```

```python
analyzer(max_depth=1)
```
